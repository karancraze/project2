```python
from random import seed
from random import randint
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from math import sqrt
from sklearn.metrics import mean_squared_error


# generate examples of random integers and their sum
def random_sum_pairs(n_examples, n_numbers, largest):
	X, y,list_100 = list(), list(),list()

	for i in range(n_examples):
		in_pattern = [randint(1,largest) for _ in range(n_numbers)]
		#input_seq.append(in_pattern)
		list_100 = in_pattern
		out_pattern = in_pattern
		X.append(in_pattern)
		y.append(out_pattern)
	# format as NumPy arrays
	list_100 = X   
	X,y = array(X), array(y)
	# normalize
	X = X.astype('float') / float(largest * n_numbers)
	y = y.astype('float') / float(largest * n_numbers)
	return X, y,list_100

# invert normalization
def invert(value, n_numbers, largest):
	return round(value * float(largest * n_numbers))

# generate training data
seed(1)
n_examples = 2000
n_numbers = 1
largest = 1000
# define LSTM configuration
n_batch = 1
n_epoch = 1000
```

```python
# create LSTM
model = Sequential()
model.add(LSTM(10, input_shape=(n_numbers, 1)))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
# train LSTM
```

# n_examples=100, n_numbers =2, largest=100

```python
for _ in range(n_epoch):
    X, y,list_100 = random_sum_pairs(n_examples, n_numbers, largest)
    X = X.reshape(n_examples, n_numbers, 1)
    
    #model.fit(X, y, epochs=1, batch_size=n_batch, verbose=2)

```

```python
model.fit(X, y, epochs=n_epoch, batch_size=n_batch, verbose=2)
```

```python

result = model.predict(X, batch_size=n_batch, verbose=0)
```

```python
# calculate error
expected = [round(float(largest * n_numbers) * x[0],2) for x in y]
predicted = [round(float(largest * n_numbers) * x) for x in result[:,0]]
rmse = sqrt(mean_squared_error(expected, predicted))
f = open(str(n_examples)+'_numbers.txt','w')
print('RMSE: %f' % rmse)
f.write('RMSE: %f' % rmse +"\n")
total = len(expected)
generatedErrors = 0

# show some examples
for i in range(n_examples):
	error = expected[i] - predicted[i]
	if(error != 0):
		generatedErrors = generatedErrors +1 
	print('Expected=%d, Predicted=%d (err=%d)' % (expected[i], predicted[i], error))
	f.write('Expected=%d, Predicted=%d (err=%d)' % (expected[i], predicted[i], error) + "\n")
    
print("generatedErrors : ",generatedErrors) 
f.write("generatedErrors : "+ str(generatedErrors) + "\n")
Accuracy = Accuracy = (total - generatedErrors) * 100/total
Accuracy = "Accuracy :"+ str(Accuracy) + "%"
print (Accuracy)
f.write("Accuracy :" + str(Accuracy) + "\n")
f.close()
```
